```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
total=len(df)
online=df[df['Has Online delivery']=='Yes']
percentage=(len(online)/total)*100
print("Percentage of restaurants offering online delivery:", round(percentage, 2), "%")
```

```python
avg_online=df[df['Has Online delivery']=='Yes']['Aggregate rating'].mean()
avg_ofline=df[df['Has Online delivery']=='No']['Aggregate rating'].mean()
print("Average rating (With Online Delivery):", round(avg_online, 2))
print("Average rating (Without Online Delivery):", round(avg_ofline, 2))
```

```python
labels = ['Online Delivery', 'No Online Delivery']
values = [avg_online, avg_ofline]

plt.bar(labels, values)
plt.title("Average Rating Comparison")
plt.xlabel("Type")
plt.ylabel("Average Rating")
plt.show()
```
