```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

df = pd.read_csv('data.csv')
df.head()
```

```python
rating_counts = df['Aggregate rating'].value_counts().sort_index()
print("Rating distribution:\n", rating_counts)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=20, kde=False, color='orange', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
average_votes = df['Votes'].mean()
print(f"The average number of votes received by restaurants is: {average_votes:.2f}")
```

```python

bins = [0, 50, 100, 500, 1000, 5000, 10000, df['Votes'].max()]
labels = ['0-50','51-100','101-500','501-1000','1001-5000','5001-10000','10000+']

df['Vote_Range'] = pd.cut(df['Votes'], bins=bins, labels=labels)

plt.figure(figsize=(8,5))
sns.countplot(x='Vote_Range', data=df, palette='pastel', order=labels)
plt.title('Number of Restaurants by Vote Range')
plt.xlabel('Vote Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```
