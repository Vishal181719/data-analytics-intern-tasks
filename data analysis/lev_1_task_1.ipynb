```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from collections import Counter
```

```python


df = pd.read_csv('data.csv', encoding='latin-1')
df.head()
```

```python
df['Cuisines'].head()
```

```python
cuisine_series=df['Cuisines'].dropna().str.split(',')
```

```python
counter = Counter()

for cuisines in cuisine_series:
    counter.update(cuisines)
```

```python
top3=counter.most_common(3)
print(top3)
```

```python
total = len(df)

for cuisines, count in top3:
    percentage = (count / total) * 100
    print(cuisines, ":", round(percentage,2), "%")
```

```python
names = [cuisine for cuisine, count in top3]
counts = [count for cuisine, count in top3]

plt.bar(names, counts)
plt.title("Top 3 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```
